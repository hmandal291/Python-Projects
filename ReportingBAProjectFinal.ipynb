```python
url ='https://raw.githubusercontent.com/IntuitiveWebSolutions/ReportingBAProject/master/data/policies.csv'
policies = pd.read_csv(url,sep=',')
policies.head(20)
```

```python
url ='https://raw.githubusercontent.com/IntuitiveWebSolutions/ReportingBAProject/master/data/policyholders.csv'
policyholders = pd.read_csv(url,sep=',')
policyholders.head(5)
```

```python
url ='https://raw.githubusercontent.com/IntuitiveWebSolutions/ReportingBAProject/master/data/revisions.csv'
revisions = pd.read_csv(url,sep=',')
revisions['revisionDate']=pd.to_datetime(revisions['revisionDate'])
revisions['cancelDate']=pd.to_datetime(revisions['cancelDate'])
revisions.head(5)
```

Merging policies, Revisions and PolicyHolders

```python
pore=pd.merge(policies, revisions, on='policyId', how='right')
Finamerge=pd.merge(policyholders,pore, on='revisionId', how='right')
Finamerge
```

Joining all the Policy Holders together in the same name section column

```python
Finamerge['key1'] = (Finamerge['revisionId'] != Finamerge['revisionId'].shift(1)).astype(int).cumsum()
Finamerge['policyholderName']=Finamerge.groupby('key1', sort=False)['policyholderName'].apply(','.join)
Finamerge=Finamerge.dropna(subset = ['policyholderName'])
Finamerge.head(5)
```

SORTING BY policyholderName

```python
Finamerge=Finamerge.sort_values(['policyholderName'])
Finamerge = Finamerge.reset_index(drop=True)
Finamerge.head(5)
```

SORTING BY policyholderName, revisionDate, commitDate

```python
Finamerge['commitDate']=pd.to_datetime(Finamerge['commitDate'])
Finamerge['dateArchived']=pd.to_datetime(Finamerge['dateArchived'])
Finamerge['revisionDate']=pd.to_datetime(Finamerge['revisionDate'])
Finamerge['effectiveDate']=pd.to_datetime(Finamerge['effectiveDate'])
Finamerge=Finamerge.sort_values(['policyholderName','revisionDate','commitDate'])
Finamerge.head(10)
```

```python
Finamerge = Finamerge.reset_index(drop=True)
```

```python
def Transaction_Type (row):

        if row['policyStatus'] == 'Canceled' :
          return 'Canceled'
        if (row['policyStatus'] == 'Active') & (row['revisionState'] == 'committed') & (row['revisionDate']==row['effectiveDate']) :
            return 'New'
        return 'Endorsement'
```

```python
Finamerge['Transaction_Type'] = Finamerge.apply (lambda row: Transaction_Type (row),axis=1)
```

```python
Finamerge.head(10)
```

```python
#Loop to set Reinstated in the Transaction_Type column
for index, row in Finamerge.iterrows():
    if row['policyStatus'] == 'Canceled':
       if Finamerge.iloc[index,3]== Finamerge.iloc[index+1,3] :
        Finamerge.iloc[index+1,14]='Reinstated'
```

# Creating a new column called Change_in_Premium

```python
idx = Finamerge.index
Finamerge['Change_in_Premium'] = Finamerge['writtenPremium'].diff()
mask =( Finamerge.policyNumber != Finamerge.policyNumber.shift(1))
Finamerge['Change_in_Premium'][mask] = iterationRow()
Finamerge.reindex(idx)
```

```python
Finamerge['Change_in_Premium']=Finamerge['Change_in_Premium'].fillna(1)
```

```python
def iterationRow():
    conditions=[(Finamerge['revisionState'] == 'archived') & (Finamerge['Transaction_Type']=='Endorsement'), 
            (Finamerge['revisionState'] == 'committed') & (Finamerge['Transaction_Type']=='New')]
    choices=[- Finamerge['writtenPremium'],Finamerge['writtenPremium'] ]
    return (np.select(conditions, choices ))
```
